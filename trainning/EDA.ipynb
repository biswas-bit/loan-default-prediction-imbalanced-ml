Imputing the numerical data with its median and categorical with unknowm

```python
def numerical_data_missing(df):
    missing_percent = df.isnull().mean()*100
    cols_missing_10_60 = missing_percent[(missing > 10) & (missing <=60)].index.tolist()
    numeric = []
    for col in cols_missing_10_60:
        if df[col].dtype in ["int64","float64"] and df[col].nunique()!=2:
            numeric.append(col)
    return numeric
    
def categorical_data_missing(df):
    missing_percent = df.isnull().mean() * 100
    cols_missing_to_10_60 = missing_percent[(missing_percent > 10) & (missing_percent <= 60)].index.tolist()
    categorical = []
    for col in cols_missing_to_10_60:
        if df[col].dtype in ["object", "category"]:
            categorical.append(col)

    return categorical

def binary_data_missing(df):
    missing_percent = df.isnull().mean() * 100
    cols_missing_to_10_60 = missing_percent[(missing_percent > 10) & (missing_percent <= 60)].index.tolist()
    binary = []
    for col in cols_missing_to_10_60:
        if df[col].nunique() == 2:
            binary.append(col)

    return binary
```

```python
# working on numerical data (10-60)
df[numerical_data_missing(df)] = df[numerical_data_missing(df)].fillna(df[numerical_data_missing(df)].median())
```

```python
# working on categorical data (10-60)
df[categorical_data_missing(df)] = df[categorical_data_missing(df)].fillna("unknown",)
```

```python
# working on binary data (10-60)
df[binary_data_missing(df)] = df[binary_data_missing(df)].apply(lambda x: x.fillna(x.mode()[0]))
```

working on missing > 10 replacing categorical with mode and numerical with median

```python
def numerical_small(df):
    missing_percent = df.isnull().mean()*100
    cols_missing_10_60 = missing_percent[missing<10].index.tolist()
    numeric = []
    for col in cols_missing_10_60:
        if df[col].dtype in ["int64","float64"] and df[col].nunique()!=2:
            numeric.append(col)
    return numeric
    
def categorical_small(df):
    missing_percent = df.isnull().mean() * 100
    cols_missing_to_10_60 = missing_percent[missing<10].index.tolist()
    categorical = []
    for col in cols_missing_to_10_60:
        if df[col].dtype in ["object", "category"]:
            categorical.append(col)

    return categorical

def binary_missing(df):
    missing_percent = df.isnull().mean() * 100
    cols_missing_to_10_60 = missing_percent[missing<10].index.tolist()
    binary = []
    for col in cols_missing_to_10_60:
        if df[col].nunique() == 2:
            binary.append(col)

    return binary
```

```python
# working on numerical data (<10)
df[numerical_small(df)] = df[numerical_small(df)].fillna(df[numerical_small(df)].median())
```

```python
# working on categorical data (<10)
df[categorical_small(df)] = df[categorical_small(df)].fillna( df[categorical_small(df)].mode())
```

```python
# working on binary data (<10)
df[binary_missing(df)] = df[binary_missing(df)].apply(lambda x: x.fillna(x.mode()[0]))
```

Correlation with target 

```python
def numerical_data(df):
    numeric = []
    for col in df.columns:
        if df[col].dtype in ["int64","float64"] and df[col].nunique()!=2:
            numeric.append(col)
    return numeric
    
```

```python
target = 'TARGET'
numeric_cols = numerical_data(df)
corr_with_target = (
    df[numeric_cols + [target]]
    .corr()[target]
    .drop(target)
    .sort_values(key=abs, ascending=False)
)
corr_with_target.head(50)
```

```python
# categorizing features by corellation strength
corr_df = corr_with_target.reset_index()
corr_df.columns = ['feature', 'corelation']
corr_df['abs_corr'] = corr_df['corelation'].abs()
corr_df.head(50)
```

```python
# visulizing top corellated Features 
top_corr = corr_df.sort_values('abs_corr',ascending=False).head(20)
plt.figure(figsize=(8,6))
sns.barplot(
    data=top_corr,
    x='corelation',
    y='feature'
)
plt.title("top 20 Numerical Features Corelated with Target")
plt.tight_layout()
plt.show()
```

```python
# categorical Features 
def categorical_data(df):
    categorical=[]
    for cols in df.columns:
        if df[cols].dtype in ["object"]:
            categorical.append(cols)
    return  categorical
        

cat_cols = categorical_data(df)
cat_target_mean = {}
for col in cat_cols:
    cat_target_mean[col]= df.groupby(col)[target].mean().sort_values(ascending=False)
cat_target_mean['NAME_INCOME_TYPE'].head(100)
```

```python
def binary(df):
    binary = []
    for col in df.columns:
        if df[col].nunique() == 2:
            binary.append(col)

    return binary

binary_cols = binary(df)

binary_analysis = []

for col in binary_cols:
    rate_0 = df[df[col] == 0][target].mean()
    rate_1 = df[df[col] == 1][target].mean()
    binary_analysis.append({
        'feature': col,
        'default_rate_0': rate_0,
        'default_rate_1': rate_1,
        'difference': rate_1 - rate_0
    })

binary_df = pd.DataFrame(binary_analysis).sort_values(
    by='difference', key=abs, ascending=False
)

binary_df.head(10)
```

Multicollinearity (Corellation Heatmap + VIF)

```python
num_cols = numerical_data(df)
# removing constant feature before corelation cause it was causing the problem of nan due to the no variance
num_cols = [c for c in num_cols if df[c].nunique()>1]
corr_matrix = df[num_cols].dropna().corr()
corr_matrix
```

```python
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
plt.figure(figsize=(14,12))
sns.heatmap(
    corr_matrix,
    mask=mask,
    cmap="coolwarm",
    center =0,
    linewidth=0.5
)
plt.title("corelation Heatmap (numerical Features)")
plt.show()
```
